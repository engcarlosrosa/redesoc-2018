O atributo _iloc_ de um _DataFrame_ permite usar slices baseados em índices numéricos. Vale a mesma notação usada no numpy. Para saber mais veja [aqui](http://pandas.pydata.org/pandas-docs/stable/indexing.html)

```python
dados.iloc[0:10, 0:8]
```

Pode-se usar dois pontos *:*  em uma das dimensões para selecionar todas

```python
dados.iloc[:, 6]
```

Ou então utilizar uma tupla para especificar apenas um conjunto limitado de células

```python
dados.iloc[0:10, (0, 6, 9)]
```

Pode-se especificar uma condição para selecionar apenas as linhas em que aquela condição é válida

```python
dados[dados.DEFAULT==0]
```

Em cima da condição especificada, podemos ainda selecionar apenas uma coluna

```python
pd.DataFrame.plot(kind='bar', data=dados.DEFAULT.value_counts(), color=('blue', 'red'), legend=False)
```

```python
pd.DataFrame.plot(kind='pie', data=dados.DEFAULT.value_counts(), colors=('blue', 'red'), legend=False)
```

```python
pd.DataFrame.plot(kind='bar', data=dados.EDUCACAO.value_counts(), color=('yellow', 'salmon', 'blue', 'red', 'green'), legend=False)
```

```python
pd.DataFrame.plot(kind='pie', data=dados.EDUCACAO.value_counts(), colors=('yellow', 'salmon', 'blue', 'red', 'green'), autopct="%.2f", legend=False)
```

```python
pd.DataFrame.plot(kind='bar', data=pd.crosstab(dados.EDUCACAO, dados.DEFAULT), color=('yellow', 'salmon'), stacked=True)
```
